```python
import numpy as np
import h5py
from PIL import Image
import matplotlib.pyplot as plt
```

```python
filepath = r"/honey/nmep/medium-imagenet-96.hdf5"
ds = h5py.File(filepath, "r")
samples = []
labels = ["train", "val", "test"]

for i in range(15):
    idx = np.random.randint(0, len(ds))
    sample = (Image.fromarray(np.transpose(ds[f"images-{labels[i % 3]}"][idx], (2, 1, 0))), labels[i % 3]) 
    samples.append(sample)
```

```python
plt.figure(figsize=(8, 6))
for i, (image, label) in enumerate(samples):
    axs = plt.subplot(3, 5, i+1)
    axs.set_title(label)
    plt.imshow(image)
```
